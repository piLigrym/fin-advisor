```python
import pandas as pd
import numpy as np

df = pd.read_csv('Family Income and Expenditure.csv')
df.head()
```

```python
df.columns
```

```python
# New data frame only finance
# Ignore all food subcategory

financial_columns = [
    'Total Household Income', 'Main Source of Income','Total Food Expenditure', 'Restaurant and hotels Expenditure',
    'Alcoholic Beverages Expenditure', 'Tobacco Expenditure', 'Clothing, Footwear and Other Wear Expenditure', 
    'Housing and water Expenditure','Medical Care Expenditure', 
    'Transportation Expenditure','Communication Expenditure', 'Education Expenditure', 
    'Miscellaneous Goods and Services Expenditure', 'Special Occasions Expenditure',
    'Crop Farming and Gardening expenses'
]
df_f = df[financial_columns]
df_f.head()
```

```python
# Rename columns
new_column_names = {
    'Total Household Income': 'total_income',
    'Main Source of Income': 'income_source',
    'Total Food Expenditure': 'food',
    'Restaurant and hotels Expenditure': 'hospitality',
    'Alcoholic Beverages Expenditure': 'alcohol',
    'Tobacco Expenditure': 'tobacco',
    'Clothing, Footwear and Other Wear Expenditure': 'clothing',
    'Housing and water Expenditure': 'public_utilities',
    'Medical Care Expenditure': 'medical',
    'Transportation Expenditure': 'transport',
    'Communication Expenditure': 'communication',
    'Education Expenditure': 'education',
    'Miscellaneous Goods and Services Expenditure': 'others',
    'Special Occasions Expenditure': 'special',
    'Crop Farming and Gardening expenses': 'gardening'
}
df_f = df_f.rename(columns=new_column_names)
df_f.head()
```

```python
unique_sources_of_income = df_f['income_source'].unique()
unique_sources_of_income
```

```python
# Create new column saving
df_f['saving'] = df_f['total_income'] - (
    df_f['food'] +
    df_f['hospitality'] +
    df_f['alcohol'] +
    df_f['tobacco'] +
    df_f['clothing'] +
    df_f['public_utilities'] +
    df_f['medical'] +
    df_f['transport'] +
    df_f['communication'] +
    df_f['education'] +
    df_f['others'] +
    df_f['special'] +
    df_f['gardening']
)
 
negative_saving_count = (df_f['saving'] < 0).sum()
negative_saving_count
```

```python
negative_saving_rows = df_f[df_f['saving'] < 0]
negative_saving_rows.head()
```

```python
# Delete all negative saving
df_f = df_f.loc[df_f['saving'] >= 0].reset_index(drop=True)
print(df_f.shape)
df_f.head()
```

```python
# Create new column total_expenses
df_f['total_expenses'] = (
    df_f['food'] +
    df_f['hospitality'] +
    df_f['alcohol'] +
    df_f['tobacco'] +
    df_f['clothing'] +
    df_f['public_utilities'] +
    df_f['medical'] +
    df_f['transport'] +
    df_f['communication'] +
    df_f['education'] +
    df_f['others'] +
    df_f['special'] +
    df_f['gardening']
)

total_expenses = df_f.pop('total_expenses')
df_f.insert(1, 'total_expenses', total_expenses)
df_f.head()
```

```python
# Make saving more random
random_addition = np.random.uniform(0, 0.3, df_f.shape[0]) * df_f['saving']

df_f['saving'] += random_addition

df_f['saving'] = df_f['saving'].round(0).astype(int)
df_f.head()
```

```python
# Generation of minimum investment
df_f['min_invest'] = (
    (df_f['saving'] / df_f['total_income'] * 0.2 + df_f['total_expenses'] / df_f['total_income'] * 0.1) * df_f['saving']
).round(0).astype(int)
df_f.head()
```

```python
df_f.to_csv('financial_data.csv', index=False)
```
